```python
#import dependencies
import matplotlib.pyplot as plt
```

```python
#set the x axis to a list of strings for each month
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

#set the y_axis to a list of floats as the total fare in the US dollars accumulated for each month
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
#create the plot
plt.plot(x_axis, y_axis)
```

```python
#create the plot with ax.plot()
fig, ax = plt.subplots()
ax.plot(x_axis,y_axis)
```

```python
#create the plot with ax.plot()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis,y_axis)
```

```python
#create the plot with ax.plot()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
#create the plot and add a label for the legent
plt.plot(x_axis, y_axis, label='Boston')

#Create labels for the x and y axes
plt.xlabel("Date")
plt.ylabel("Fare($)")

#set the y limit between 0 and 45
plt.ylim(0,45)

#Create a title
plt.title("Pyber Fare by Month")

#Add the legend
plt.legend()
```

```python
#create the plot and add a label for the legent
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')

#Create labels for the x and y axes
plt.xlabel("Date")
plt.ylabel("Fare($)")

#set the y limit between 0 and 45
plt.ylim(0,45)

#Create a title
plt.title("Pyber Fare by Month")

#Add a grid
plt.grid()

#Add the legend
plt.legend()
```

```python
#revise chart using object oriented approach
fig, ax = plt.subplots()

#create the plot and add a label for the legend
ax.plot(x_axis, y_axis, marker='d', color="green", linewidth=2, label='Boston')

#Create labels for the x and y axes
ax.set_xlabel("Date")
ax.set_ylabel("Fare($)")

#set the y limit between 0 and 45
ax.set_ylim(0,45)

#Create a title
ax.set_title("Pyber Fare by Month")

#Add a grid
ax.grid()

#Add the legend
ax.legend()
```

```python
#create the plot
plt.bar(x_axis, y_axis)
```

```python
#create the plot
plt.bar(x_axis, y_axis, color="green", label='Boston')

#create labels for the x and y axes
plt.xlabel("Date")
plt.ylabel("Fare($)")

#Create a title
plt.title("Pyber Fare by Month")

#Add the legend
plt.legend()
```

```python
#create the plot
plt.barh(x_axis, y_axis, color="magenta", label='Boston')
plt.gca().invert_yaxis()

plt.xlabel("Fare($)")
plt.ylabel("Date")

plt.legend()
```

```python
#Create the plot with the ax.plot()
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis)
```

```python
#Create the plot with the ax.plot()
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, color='cyan', label="Chicago")
ax.invert_yaxis()

ax.set_title("Pyber Fare by Month")

ax.legend()

ax.set_ylabel("Date")
ax.set_xlabel("Fare ($)")



```

```python
plt.scatter(y_axis, x_axis, color="red", label='Chicago')
plt.legend()
plt.title("Pyber Fare by Month")
plt.ylabel("Date")
plt.xlabel("Fare ($)")
plt.xlim(0,45)
plt.gca().invert_yaxis()
```

```python
plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)
    
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
plt.scatter(x_axis, y_axis, s=[i * 3 for i in y_axis])
```

```python
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, s=[i * 5 for i in y_axis], c='skyblue', edgecolors='black', label="Boston")
ax.legend()

ax.set_title("Pyber Fare by Month")
ax.set_ylabel("Date")
ax.set_xlabel("Fare ($)")
ax.invert_yaxis()
ax.set_xlim(0,45)
```

```python

plt.pie(y_axis, labels=x_axis)
```

```python
plt.subplots(figsize=(8, 8))
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
```

```python
#Assign 12 colors, one for each month
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

```

```python
fig, ax = plt.subplots(figsize=(8, 8))
explode_values = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
ax.pie(y_axis, 
       labels=x_axis,
      autopct='%1.1f%%',
      explode=explode_values,
      shadow=True,
      colors=colors,
      startangle=94,
      counterclock=False)
plt.show()
```
